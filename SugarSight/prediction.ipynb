```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,recall_score,precision_score,f1_score
import joblib
import warnings
warnings.filterwarnings("ignore")
sns.set_style('darkgrid')
```

```python
diabetes_data = pd.read_csv('diabetes.csv')
diabetes_data.head()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
outcome_counts = diabetes_data['Outcome'].value_counts()
plt.bar(['Non Diabetic', 'Diabetic'], outcome_counts.values, color=['skyblue', 'salmon'])
plt.title('Distribution of Diabetes Outcome', fontsize=16, fontweight='bold')
plt.ylabel('count')
plt.show()

plt.subplot(1, 2, 2)
plt.pie(outcome_counts.values,
        labels=['Non Diabetic', 'Diabetic'],
        autopct='%1.1f%%',
        colors=['skyblue', 'salmon'],
        startangle=80)
plt.title('Proportion of Diabetes Outcome', fontsize=16, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,7))
correlation = diabetes_data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, square=True, linewidths=1, fmt=".2f")
plt.title('Correlation Heatmap of Diabetes Dataset', fontsize=16, pad=18, fontweight='bold')
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 12))
features = diabetes_data.columns[:-1]

for idx, feature in enumerate(features):
    row = idx // 3
    col = idx % 3
    for outcome in [0, 1]:
        data = diabetes_data[diabetes_data['Outcome'] == outcome][feature]
        axes[row, col].hist(data, bins=15, alpha=0.5, 
                            label=f'Outcome {outcome}', 
                            color=['blue', 'orange'][outcome])
    axes[row, col].set_title(f'Distribution of {feature}', fontweight='bold')
    axes[row, col].set_xlabel('Value')
    axes[row, col].legend()

# Hide the last empty subplot
axes[2, 2].axis('off')

plt.tight_layout()
plt.show()
```

```python
X = diabetes_data.drop(columns='Outcome', axis=1)
Y = diabetes_data['Outcome']

print(X)
print(Y)
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(X)
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, Y, test_size=0.2, stratify=Y, random_state=2)
print(x.shape, x_train.shape, x_test.shape)
```

```python
classifier = svm.SVC(kernel='linear', probability=True, class_weight='balanced')
classifier.fit(x_train, y_train)
```

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
training_data_accuracy
```

```python
precision = precision_score(y_train, x_train_prediction)
recall = recall_score(y_train, x_train_prediction)
f1 = f1_score(y_train, x_train_prediction)
print(f'Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
```

```python
cm_svm = confusion_matrix(y_train, x_train_prediction)
cm_svm
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', square=True, cbar=False)
xticklabels = ['Predicted Non-Diabetic', 'Predicted Diabetic']
yticklabels = ['Actual Non-Diabetic', 'Actual Diabetic']
plt.title('Confusion Matrix - SVM Classifier', fontsize=16, fontweight='bold')
plt.ylabel('Actual', fontsize=12)
plt.xlabel('Predicted', fontsize=12)
plt.show()

```

```python
rf_classifier = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=3,
    max_features='sqrt',
    class_weight='balanced'  # if classes imbalanced
)
rf_classifier.fit(x_train, y_train)
```

```python
x_train_prediction_rf = rf_classifier.predict(x_train)
```

```python
training_data_accuracy_rf = accuracy_score(y_train, x_train_prediction_rf)
training_data_accuracy_rf
```

```python
x_test_prediction_rf = rf_classifier.predict(x_test)
test_data_accuracy_rf = accuracy_score(y_test, x_test_prediction_rf)
```

```python
test_data_accuracy_rf
```

```python
cm_rf = confusion_matrix(y_test, x_test_prediction_rf)
print(cm_rf)
```
